## Предобработка данных

### Обработка пропусков

Посчитаем количество пропусков во всех столбцах датасета:

#### Столбцы 'ceiling_height' и  'floors_total' 

<a id='ceiling'></a>
Откуда могли взяться пропуски? Возможно, продавец просто не заполнил данный параметр, посчитав его несущественным. Такое могло случиться, если высота потолков обычная и ничем не примечательная. В противном случае, это было бы преимуществом и его бы точно указали. Значит, логично будет заполнить пропущенные значения каим-то типичным значением.

Посчитаем долю пропусков и посмотрим на распределение имеющихся значений:

```python
print(f'Доля пропусков в столбце "ceiling_height": {df["ceiling_height"].isna().mean():.0%}')
df['ceiling_height'].describe()
```

Самая распространенная высота потолков - от 2.52 до 2.80 м. Медиана - 2.65 м (вполне реалистично).

Посмотрим на распределение с помощью гистограммы:

```python
df['ceiling_height'].hist(bins=100, range=(2, 4), figsize=(12,4))
plt.xlabel('Высота потолка, м')
plt.ylabel('Частотность')
plt.show()
```

Если мы просто заменим все пропуски (39% данных столбца) одной медианой - 2.65, то в результате получим очень высокий столбик на отметке 2.65. Но нам желательно сохранить форму распределения. Высота потолков зависит от типа(серии) дома. Серию дома мы можем косвенно определить по количеству этажей + количеству комнат + общей площади. 

Поэтому пропуски можно заменить медианой по столбцу 'ceiling_height' в зависимости от количества этажей, количества комнат и общей площади. Однако для этого сначала необходимо заполнить пропуски в столбце 'floors_total' и категоризировать столбцы,  которые собираемся группировать ('rooms', 'floors_total', 'total_area').

Заменим пропуски в **столбце 'floors_total'**. Причиной их появления могли стать забывчивость или невнимательность продавца, который не заполнил соответствующее поле. Количество пропусков незначительное - 86 шт. или менее 1%. Можно было бы просто удалить строки с пропусками, но жаль терять данные из оставшегося 21 столбца.

Общее количество этажей должно быть больше либо равно занчению в столбце 'floor' (там пропусков нет). Если мы приравняем этажность дома к этажу квартиры, то получится, что квартира находится на последнем этаже, а это существенный признак, влияющий на цену. Поэтому предположим, что квартира, где пропущена этажность дома, находиться на предпоследнем этаже и заменим пропуски на (этаж квартиры + 1).

```python
df['floors_total'] = df['floors_total'].fillna(df['floor'] + 1)
df['floors_total'].isna().sum()
```

Поскольку мы собираемся в дальнейшем группировать данные по столбцу 'floors_total' (чтобы более обоснованно заменить пропуски в столбце 'ceiling_height'), то необходимо поменять тип данных на целочисленный и провести категоризацию.

```python
df['floors_total'] = df['floors_total'].astype('int')
df['floors_total'].dtypes
```

Тип данных поменяли, приступим к категоризации столбца 'floors_total'. Для этого сначала посмотрим на распределение его значений.

```python
df['floors_total'].describe()
```

```python
df['floors_total'].hist(range=(1,30), bins=30, figsize=(12,4))
plt.xlabel('Количество этажей')
plt.ylabel('Частотность')
plt.show()
```

Выделим следующие категории этажности дома:
- 1-5 этажей
- 6-10 этажей
- более 10 этажей

Создадим столбец и запишем в него категории этажности, для этого напишем функцию:

```python
def fl_total_category(fl_total):
    '''
    функция будет принимать в качестве аргумента этаж 
    и возвращать категорию этажности
    '''
    if fl_total < 6:
        return '1-5 этажей'
    if fl_total < 11:
        return '6-10 этажей'
    return 'более 10 этажей'

# проверим работоспособность функции
fl_total_category(15)
```

Создадим новый столбец, поместим в него результат применения функции к столбцу 'floors_total', проверим результат:

```python
df['fl_total_category'] = df['floors_total'].apply(fl_total_category)
df.head(3)
```

Теперь займемся категоризацией **столбца 'total_area'**. Посмотрим на распределение.

```python
df['total_area'].describe()
```

Чтобы удобнее было выбирать границы категорий, построим гистограмму:

```python
df['total_area'].hist(bins=110, range=(10, 120), figsize=(12,4))
plt.xlabel('Общая площадь, м2')
plt.ylabel('Частотность')
plt.show()
```

Выделим следующие категории общей площади:
- менее 30 м2
- от 30 до 40 м2
- от 40 до 50 м2
- от 50 до 70 м2
- более 70 м2

Создадим новый столбец и запишем туда категории с помощью функции:

```python
def area_category(area):
    '''
    функция принимает в качестве аргумента общую площадь 
    и возвращает категорию площади
    '''
    if area < 30:
        return 'до 30 м2'
    elif area < 40:
        return '30-40 м2'
    elif area < 50:
        return '40-50 м2'
    elif area < 70:
        return '50-70 м2'
    return 'от 70 м2'

# проверим работоспособность функции
print(area_category(100))
```

Применим функцию к столбцу 'total_area', результаты запишем в новый столбец, проверим результат:

```python
df['area_category'] = df['total_area'].apply(area_category)
df.head(3)
```

Для категоризации **столбца 'rooms'** посмотрим на распределение значений:

```python
df.hist('rooms', bins=8, range=(0,7), figsize=(10,4))
plt.xlabel('Количество комнат')
plt.ylabel('Частотность')
plt.show()

df['rooms'].describe()
```

Что за квартиры с 0 комнат? Может, это студии? Проверим: <a id='rooms'></a>

```python
print('Количество квартир, где количество комнт 0:', len(df.query('rooms == 0')))
print('Количество квартир - студий:', len(df.query('studio == 1')))
print('Количество квартир - студий, у которых 0 комнат:', len(df.query('studio == True and rooms == 0')))
```

Оказалось, что не только у студий указано 0 комнат. (И в датасете, кстати, есть студии с другим количеством комнат). Может, остальные 59 квартир - это квартиры с открытой планировкой? Проверим:

```python
print('Количество квартир со свободной планировкой, у которых 0 комнат:', \
      len(df[(df['rooms'] == 0) & (df['studio'] == 0) & (df['open_plan'] == 1)]))
```

Итак, значение 0 комнат характерно либо для студий либо для квартир с открытой планировкой. Значит, это не ошибка. 

Но в любом случае объектов с 0 комнат слишком мало, чтобы выделять их в отдельную категорию, объединим их с 1-комнатными. Поэтому для количества комнат определим категории следующим образом:
- 1 комната
- 2 комнаты
- 3 комнаты
- более 3 комнат

Приступим к категоризации количества комнат:

```python
def rooms_category (rooms):
    '''
    функция принимает в качестве аргумента количество комнат 
    и возвращающую категорию
    '''
    if rooms < 2:
        return '1 комн'
    elif rooms < 3:
        return '2 комн'
    elif rooms < 4:
        return '3 комн'
    return 'многокомн'

# проверим работоспособность функции
rooms_category (5)
```

Применим функцию к столбцу 'rooms', результаты запишем в новый столбец, проверим результат:

```python
df['rooms_category'] = df['rooms'].apply(rooms_category)
df.head(3)
```

Теперь можем группировать датафрейм по 3 м столбцам ('fl_total_category', 'area_category', 'rooms_category') и заменять пропуски в столбце 'ceiling_height' медианами для разных групп.

```python
df['ceiling_height'] = df['ceiling_height']\
    .fillna(df.groupby(['fl_total_category', 'area_category', 'rooms_category'])['ceiling_height']\
    .transform('median'))
df['ceiling_height'].isna().sum()
```

Остались незаполненными 3 значения, судя по всему, для какой-то комбинации признаков не нашлось значений в датасете. Заполним оставшиеся пропуски просто медианой по всему столбцу.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['ceiling_height'].isna().sum()
```

Итак, мы заполнили прпопуски в столбце 'ceiling_height' (39% данных) медианами, рассчитанными с учетом группировки по 3м характеристикам:
- общей площади, 
- этажности дома, 
- количества комнат.

Параллельно заполнили пропуски в столбце 'floors_total' (менее 1% данных), предположив, что соответствующие квартиры находятся на предпоследнем этаже. Также изменили тип данных в данном столбце с дробного на целочисленный.

Кроме того, попутно категоризировали данные в столбцах 'floors_total', 'total_area', 'rooms', добавив к датасету новые столбцы - 'fl_total_category', 'area_category', 'rooms_category'.

Для этого будем группировать данные по каждому признаку и сравнивать медианные цены за м2 по каждой группе. Но для группировки нам сначала необходимо категоризировать столбец 'cityCenters_nearest' (все остальные столбцы с факторами уже категоризированы ранее).

Для столбца 'cityCenters_nearest' выделим следующие категории:

- до 5 км
- от 5 до 10 км
- от 10 до 15 км
- от 15 до 20 км
- свыше 20 км

Запишем категории в новый столбец с помощью функции:

```python
def center_dist_cat(distance):
    '''
    функция принимаюет в качестве аргумента расстояние до цента 
    и возвращает категорию
    '''
    if distance <= 5:
        return 'до 5 км'
    if distance <= 10:
        return 'от 5 до 10 км'
    if distance <= 15:
        return 'от 10 до 15 км'
    if distance <= 20:
        return 'от 15 до 20 км'
    return 'свыше 20 км'

# проверим работоспособность функции
center_dist_cat(28)
```

Создадим новый столбец, запишем туда категории, применив функцию к столбцу 'cityCenters_nearest':

```python
df_days['center_dist_cat'] = df_days['cityCenters_nearest'].apply(center_dist_cat)
df_days.head(3)
```

### Расчет корреляции

Для таких факторов, как  стоимость, цена за м2, площадь, количество комнат и расстояние до центра рассчитаем коэффициент корреляции:

```python
df_days[['last_price', 'price_per_m2', 'total_area', 'rooms', 'cityCenters_nearest']].corr()
```

Таблица показывает, что макимпльная зависимость(0,78) наблюдается между общей площадью и стоимостью квартиры. Чем больше площадь квартиры, тем она дороже, при прочих равных условиях. При этом цена за м2 от общей площади не зависит (коэффициент корреляции 0,1).

Цена за м2 зависит от стоимости квартиры (коэффициент корреляции 0,65), это логично, т.к. цена за м2 = стоимость / площадь. Чем выше стоимость квартиры, тем выше цена за м2.

От количества комнат зависимость уже ниже: стоимость  квартиры зависит от количества комнат с коэффициентом корреляции 0,49.Чем больше комнат, тем выше стоимость квартиры. А вот цена за м2 от количества комнат практически не зависит (коэффициент корреляции -0,1).

От расстояния до центра зависимость отрицательная: -0,41 для цены за м2 и -0,34 для стоимости квартиры. Это означает, что чем больше расстояние от центра, тем ниже цена за м2 и стоимость квартиры, при прочих равных условиях. Корреляция ниже 0,5 считается низкой. Это говорит о том, что на стоимость квартиры и цену за м2 влияют одновременно несколько факторов.

### Сравнение выводов по квартирам в центре и общих выводов по всему городу

Количество объектов в центре слишком мало по сравнению с количеством объектов во всем датасете, чтобы сравнивать два распределения на одном графике. Поэтому воспользуемся методом describe().

Как влияет удаленность от центра, мы уже исследовали [здесь](#distance1) и [здесь](#distance2).
Проверим теперь, как влияют на цену такие факторы как количество комнат, этаж, дата размещения объявления. 


Напишем функцию для сравнения факторов:

```python
factors = ['rooms_category','floor_category','weekday','month','year']

def compare3(factor):
    '''
    функция принимает название фактора, строит 2 сводные таблицы - для всего города и центра,
    объединяет эти таблицы и строит график
    '''
    total_pivot = df_days.pivot_table(index=factor, values='price_per_m2', aggfunc='median')
    total_pivot = total_pivot.rename(columns={'price_per_m2':'price_total'})
    
    center_pivot = df_days_center.pivot_table(index=factor, values='price_per_m2', aggfunc='median')
    center_pivot = center_pivot.rename(columns={'price_per_m2':'price_center'})
    
    result_pivot = total_pivot.join(center_pivot)
    result_plot = result_pivot.plot(grid=True, style='o-', figsize=(12,4))
    
    display(result_pivot)
    display(result_plot)
```

Сравним **количество комнат**:

```python
compare3('rooms_category')
```

Чем больше комнат, тем ниже цена за м2. 

Для центра это касается всех квартир, т.к. в центре выше доля элитного жилья.


Для города в целом это правило касается жилья внутри одного класса. В выборке основная доля - это обычное жилье эконом-класса, где количество комнат обычно не превышает 3. Среди многокомнатных квартир по городу в целом доля элитного жилья больше, чем среди 3-комнтаных, поэтому цены на 3-комнатные и многокомнатные на одном уровне.

Разница цен между городом в целом и центром снижается с увеличением числа комнат: с 30 тыс.руб./м2 за однокомнатные до 13 тыс.руб./м2 за многокомнатные. Это также объясняется тем, что с увеличением числа комнат доля элитного жилья по городу в целом приближается к доле элитного жилья в центре.

Изучим влияние **этажа** на цену:

```python
compare3('floor_category')
```

Первые этажи  - самые дешевые, средние - самые дорогие, это одинаково справедливо и для центра, и для города в целом.

Разница между городом в целом и центром для первых и средних этажей одинаковая - 22 тыс.руб./м2. Для последних этажей чуть больше - 24 тыс. руб./м2, т.к. в центре на последних этажах могут встречаться пентхаусы.

Проверим влияние **дня недели** размещения объявления:

```python
compare3('weekday')
```

Интересно, что в центре квартиры дешевле в будние дни и дороже в выходные (самые дешевые - во вторник - 113 тыс.руб./м2, самые дорогие - в сб - 119 тыс.руб./м2).

В городе в целом - все наоборот: квартиры дороже в будние дни и дешевле в выходные (самые дорогие во вторник и среду - 94 тыс.руб./м2, самые дешевые - в сб - 92 тыс.руб./м2).

Таким образом, зависимость цены от дня недели размещения объявления является спорной и неоднозначной.

Проверим, как влияет **месяц** размещения объявления:

```python
compare3('month')
```

Динамика цен на объекты в центре и в городе в целом похожа и отражает сезонные колебания цен на недвижисмоть - пики в начале весны и в начале осени, спады летом и зимой.

Для объектов в центре график более резкий, колеблется от 111 до 122 тыс.руб./м2, т.е. размах составляет 11 тыс. руб.(приметно 10%). Здесь видны спады не только в июне и декабре (как на общегородском графике), но еще и в феврале, и в ноябре. Рост цен весной более выраженный и начинается на месяц раньше - в марте.

Для объектов в целом по выборке график более сглаженный, колеблется от 91 до 95 тыс.руб./м2, размах всего 4 тыс.руб. (примерно 4%). Это можно объяснить бОльшей величиной выборки + бОльшей долей дешевых объектов.
