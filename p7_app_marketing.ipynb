## Функции для расчета и анализа LTV, ROI, удержания и конверсии

### Функция для создания пользовательских профилей 

Напишем функцию, которая
- принимает данные о сессиях, заказах, событиях, затратах и список названий событий для фильтра,
- создает таблицу "profiles" с данными о первых посещениях,
- добавляет в таблицу флаги: является ли пользователь платящим и совершал ли он интересующее нас событие,
- считает количество привлеченных пользователей для каждого канала и даты,
- считает средние расходы на одного пользователя,
- добавляет стоимость привлечения пользоватля в таблицу "profiles",
- возвращает таблицу "profiles", содержащую данные о пользователях, совершивших интересующие нас события:
        - номер пользователя,
        - данные о первом посещении пользователя (время, канал, устройство, регион, дата, месяц)
        - признак, является ли пользователь платящим,
        - стоимость привлечения пользователя:

<a id='profiles'></a>

```python
def get_profiles(sessions, orders, events, ad_costs, event_names=[]):
    
    # находим параметры первых посещений для каждого пользователя:
    # время начала сессии, канал, устройство и регион для этой сесии
    profiles = (
        sessions.sort_values(by=['user_id', 'session_start'])
        .groupby('user_id')
        .agg(
            {
                'session_start': 'first',
                'channel': 'first',
                'device': 'first',
                'region': 'first',
            }
        )
        .rename(columns={'session_start': 'first_ts'})
        .reset_index()
    )

    # добавляем к полученной таблице столбцы с датой и месяцем первого посещения
    profiles['dt'] = profiles['first_ts'].dt.date
    profiles['month'] = profiles['first_ts'].astype('datetime64[M]')

    # добавляем признак платящих пользователей: если номер пользователя есть в данных с заказами,
    # то присваиваем ему флаг "True", в противном случае "False"
    profiles['payer'] = profiles['user_id'].isin(orders['user_id'].unique())

    # добавляем флаги для всех событий из event_names
    for event in event_names:
        if event in events['event_name'].unique():
            profiles[event] = profiles['user_id'].isin(
                events.query('event_name == @event')['user_id'].unique()
            )

    # считаем количество уникальных пользователей с одинаковыми источником и датой привлечения
    new_users = (
        profiles.groupby(['dt', 'channel'])
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )

    # объединяем число привлечённых пользователей и траты на рекламу по дате и каналу
    ad_costs = ad_costs.merge(new_users, on=['dt', 'channel'], how='left')

    # считаем средние расходы на одного привлеченного пользователя
    ad_costs['acquisition_cost'] = ad_costs['costs'] / ad_costs['unique_users']

    # добавляем стоимость привлечения в профили
    profiles = profiles.merge(
        ad_costs[['dt', 'channel', 'acquisition_cost']],
        on=['dt', 'channel'],
        how='left',
    )

    # стоимость привлечения органических пользователей равна нулю
    profiles['acquisition_cost'] = profiles['acquisition_cost'].fillna(0)

    return profiles
```

### Функция для расчёта удержания

Напишем функцию, которая
- принимает данные о пользователях, сессиях, дату наблюдения, горизонт анализа, список признаков группировки, условие игнорируем ли мы горизонт
- исключает пользователей, не доживших до горизонта анализа,
- добавляет к отфильтрованным пользователям даты начала сессий,
- рассчитывает лайфтаймы, получает таблицу result_raw,
- формирует таблицу удержания result_grouped с размером когорты для каждого значения признака и долей удержанных пользователей на каждый день лайфтайма в пределах горизонта анализа,
- формирует таблицу динамики удержания result_in_time, где данные группируются аналогично result_grouped, но добавляется еще и группировка по дате
- возвращает таблицы:
        - result_raw (с лайфтаймами в днях), 
        - result_grouped (с долей удержания для каждой когорты в целом),
        - result_in_time (с долей удержания для каждой когорты с разбивкой по дням привлечения).
        
<a id='retention'></a>

```python
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    
    # добавляем столбец payer в передаваемый dimensions список
    dimensions = ['payer'] + dimensions

    '''
    исключаем пользователей, не «доживших» до горизонта анализа:
    - если мы игнорируем горизонт, то последняя подходящая для анализа дата = дате наблюдения,
    - если горизонт не игнорируем, то последняя подходящая для анализа дата = (горизонт - 1 день)
    '''
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # добавляем даты начала сессий и считаем лайфтаймы
    result_raw = result_raw.merge(
        sessions[['user_id', 'session_start']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_ts']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам и лайфтаймам
    def group_by_dimensions(df, dims, horizon_days):
        '''
        - функция принимает датафрейм, признак группировки, горизонт анализа,
        - формирует сводную таблицу result, где колонки - это лайфтайм в днях, 
        строки - значения признака группировки, значения - количество уникальных пользователей,
        - рассчитывает размеры когорт для указанного признака, присоединяет к таблице result,
        - делит количество уникальных пользователей для каждого значения признака и каждого лайфтайма 
        на размер соответствующей когорты,
        - исключает из результата все лайфтаймы, превышающие горизонт анализа,
        - возвращает таблицу result, где первый столбец - размер когорты, второй и последующие - 
        доля пользователей когорты , удержанная на каждый день лайфтайма в пределах горизонта анализа.
        '''
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result_raw = result.div(result['cohort_size'], axis=0) 
        
        # оставляем столбцы лайфтайма, вошедшие в горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        
        # возвращаем размер когорт на место (там у нас после деления появились единицы)
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания (для каждой когорты в целом)
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания (для каждой когорты в разбивке по дням привлечения)
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

### Функция для расчёта конверсии

Создадим функцию, которая
- принимает данные о пользователях, покупках, дату наблюдения, горизонт анализа, список признаков группировки, условие игнорируем ли мы горизонт,
- исключает пользователей, не доживших до горизонта анализа,
- добавляет в профили пользователей дату первой покупки,
- рассчитывает лайфтаймы (от первого посещения до первой покупки), получает таблицу result_raw,
- формирует таблицу конверсии result_grouped с размером когорты для каждого значения признака и конверсией на каждый день лайфтайма в пределах горизонта анализа,
- формирует таблицу динамики конверсии result_in_time, где данные группируются аналогично result_grouped, но с дополнительной группировкой по дате,
- возвращает таблицы:
        - result_raw (с лайфтаймами в днях), 
        - result_grouped (с конверсией для каждой когорты в целом),
        - result_in_time (с конверсией для каждой когорты с разбивкой по дням привлечения).
        
<a id='conversion'></a>

```python
def get_conversion(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    last_suitable_acquisition_date = observation_date
    
    '''
    исключаем пользователей, не «доживших» до горизонта анализа:
    - если мы игнорируем горизонт, то последняя подходящая для анализа дата = дате наблюдения,
    - если горизонт не игнорируем, то последняя подходящая для анализа дата = (горизонт - 1 день)
    '''
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # определяем дату и время первой покупки для каждого пользователя
    first_purchases = (
        purchases.sort_values(by=['user_id', 'event_dt'])
        .groupby('user_id')
        .agg({'event_dt': 'first'})
        .reset_index()
    )

    # добавляем данные о первых покупках в профили
    result_raw = result_raw.merge(
        first_purchases[['user_id', 'event_dt']], on='user_id', how='left'
    )

    # рассчитываем лайфтайм для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_dt'] - result_raw['first_ts']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        '''
        - функция принимает датафрейм, признак группировки, горизонт анализа,
        - формирует сводную таблицу result, где колонки - это лайфтайм в днях, 
        строки - значения признака группировки, в ячейках - количество уникальных пользователей,
        - считает сумму с накоплением для каждой строки,
        - рассчитывает размеры когорт для указанного признака, присоединяет к таблице result,
        - делит накопленное количество пользователей для каждого значения признака и каждого лайфтайма 
        на размер соответствующей когорты,
        - исключает из результата все лайфтаймы, превышающие горизонт анализа,
        - возвращает таблицу result, где первый столбец - размер когорты, второй и последующие - 
        конверсия на каждый день лайфтайма в пределах горизонта анализа.
        '''
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        result = result.fillna(0).cumsum(axis = 1)
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        
        # делим каждую «ячейку» в строке на размер когорты и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        
        # оставляем столбцы лайфтайма, вошедшие в горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        
        # возвращаем размер когорт на место (там у нас после деления появились единицы)
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(result_raw, dimensions + ['dt'], horizon_days)

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

### Функция для расчёта LTV ("пожизненной ценности") и ROI (окупаемости)

<a id='ltv'></a>

Напишем функцию, которая
- принимает данные о пользователях, покупках, дату наблюдения, горизонт анализа, список признаков группировки, условие игнорируем ли мы горизонт,
- исключает пользователей, не доживших до горизонта анализа
- добавляет в профили пользователей даты и суммы покупок,
- рассчитывает лайфтаймы (от первого посещения до каждой покупки), получает таблицу result_raw,
- формирует таблицу выручки с размером когорты для каждого значения признака и суммой выручки с накоплением на каждый день лайфтайма в пределах горизонта анализа,
- считает LTV: делит каждую ячейку таблицы выручки на соответствующий размер когорты, получает таблицу result_grouped,
- добавляет средние издержки (CAC) по каждому признаку группировки,
- рассчитывает ROI: делит LTV на CAC, формируя таблицу roi_grouped,
- формирует таблицу динамики LTV result_in_time, где данные группируются аналогично result_grouped, но добавляется еще и группировка по дате привлечения,
- формирует таблицу динамики ROI roi_in_time, где данные группируются аналогично roi_grouped, но добавляется еще и группировка по дате привлечения,
- возвращает таблицы:
  - result_raw (с лайфтаймами в днях для каждой покупки), 
  - result_grouped (таблица LTV со средней выручкой на 1 пользователя для каждого признака группировки и лайфтайма),
  - result_in_time (динамика LTV по дате привлечения),
  - roi_grouped (таблица ROI для каждого каждого признака группировки и лайфтайма),
  - roi_in_time (динамика ROI по дате привлечения).

```python
def get_ltv(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        purchases[['user_id', 'event_dt', 'revenue']], on='user_id', how='left'
    )
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_dt'] - result_raw['first_ts']
    ).dt.days
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        '''
        - функция принимает датафрейм, признак группировки, горизонт анализа,
        - формирует сводную таблицу result, где колонки - это лайфтайм в днях, 
        строки - значения признака группировки, в ячейках - сумма выручки,
        - считает сумму с накоплением для каждой строки,
        - рассчитывает размеры когорт для указанного признака, присоединяет к таблице result,
        - делит накопленную сумму выручки для каждого значения признака и каждого лайфтайма 
        на размер соответствующей когорты (LTV),
        - исключает из результата все лайфтаймы, превышающие горизонт анализа,
        - считает средние затраты (САС) на привлечение 1 пользователя,
        - делит LTV на САС, получает ROI,
        - возвращает таблицы 
            - result, где первый столбец - размер когорты, второй и последующие - 
            накопленная выручка на каждый день лайфтайма в пределах горизонта анализа,
            - roi, где первый столбец - размер когорты, второй - сумма CAC, 
            третий и последующие - ROI на каждый день лайфтайма в пределах горизонта анализа
        '''
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='revenue', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # собираем датафрейм с данными пользователей и значениями CAC, 
        # добавляя параметры из dimensions
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()

        # считаем средний CAC по параметрам из dimensions
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost': 'mean'})
            .rename(columns={'acquisition_cost': 'cac'})
        )

        # считаем ROI: делим LTV на CAC
        roi = result.div(cac['cac'], axis=0)

        # удаляем строки с бесконечным ROI
        roi = roi[~roi['cohort_size'].isin([np.inf])]

        # восстанавливаем размеры когорт в таблице ROI
        roi['cohort_size'] = cohort_sizes

        # добавляем CAC в таблицу ROI
        roi['cac'] = cac['cac']

        # в финальной таблице оставляем размеры когорт, CAC
        # и ROI в лайфтаймы, не превышающие горизонт анализа
        roi = roi[['cohort_size', 'cac'] + list(range(horizon_days))]

        # возвращаем таблицы LTV и ROI
        return result, roi

    # получаем таблицы LTV и ROI
    result_grouped, roi_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблиц динамики убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицы динамики LTV и ROI
    result_in_time, roi_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        result_in_time,  # таблица динамики LTV
        roi_grouped,  # таблица ROI
        roi_in_time,  # таблица динамики ROI
    )
```

### Функция для сглаживания фрейма

Чтобы сделать графики динамики показателей менее "шумными", напишем функцию, которая будет принимать датафрейм и сглаживать его:

```python
def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df
```

```python
def plot_retention(retention, retention_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержание первого дня
    retention = retention.drop(columns=['cohort_size', 0])
    # в таблице динамики оставляем только нужный лайфтайм
    retention_history = retention_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'payer'])

    # в таблице графиков — два столбца и две строки, четыре ячейки
    # в первой строим кривые удержания платящих пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('payer == True').droplevel('payer').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание платящих пользователей')

    # во второй ячейке строим кривые удержания неплатящих
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey=ax1)
    retention.query('payer == False').droplevel('payer').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание неплатящих пользователей')

    # в третьей ячейке — динамика удержания платящих
    ax3 = plt.subplot(2, 2, 3)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt', 'payer']
    ]
    # фильтруем данные и строим график
    filtered_data = retention_history.query('payer == True').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания платящих пользователей на {}-й день'.format(
            horizon
        )
    )

    # в чётвертой ячейке — динамика удержания неплатящих
    ax4 = plt.subplot(2, 2, 4, sharey=ax3)
    # фильтруем данные и строим график
    filtered_data = retention_history.query('payer == False').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания неплатящих пользователей на {}-й день'.format(
            horizon
        )
    )
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_conversion(conversion, conversion_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 5))

    # исключаем размеры когорт
    conversion = conversion.drop(columns=['cohort_size'])
    # в таблице динамики оставляем только нужный лайфтайм
    conversion_history = conversion_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # первый график — кривые конверсии
    ax1 = plt.subplot(1, 2, 1)
    conversion.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Конверсия пользователей')

    # второй график — динамика конверсии
    ax2 = plt.subplot(1, 2, 2, sharey=ax1)
    columns = [
        # столбцами сводной таблицы станут все столбцы индекса, кроме даты
        name for name in conversion_history.index.names if name not in ['dt']
    ]
    filtered_data = conversion_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика конверсии пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show()
```

### Функция для визуализации LTV и ROI

Напишем функцию, которая
- принимает таблицу LTV, таблицу LTV в динамике, таблицу окупаемости, таблицу окупаемости в динамике, горизонт анализа, окно сглаживания,
- строит кривые LTV, график динамики LTV, график динамики средних затрат(CAC), кривые окупаемости(ROI), график динамики ROI:

<a id='plot_ltv'></a>

```python
def plot_ltv_roi(ltv, ltv_history, roi, roi_history, horizon, window=7):

    # задаём сетку отрисовки графиков
    plt.figure(figsize=(20, 10))

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    # в таблице динамики ltv оставляем только нужный лайфтайм
    ltv_history = ltv_history.drop(columns=['cohort_size'])[[horizon - 1]]

    # стоимость привлечения запишем в отдельный фрейм
    cac_history = roi_history[['cac']]

    # из таблицы roi исключаем размеры когорт и cac
    roi = roi.drop(columns=['cohort_size', 'cac'])
    # в таблице динамики roi оставляем только нужный лайфтайм
    roi_history = roi_history.drop(columns=['cohort_size', 'cac'])[
        [horizon - 1]
    ]

    # первый график — кривые ltv
    ax1 = plt.subplot(2, 3, 1)
    ltv.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('LTV')

    # второй график — динамика ltv
    ax2 = plt.subplot(2, 3, 2, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in ltv_history.index.names if name not in ['dt']]
    filtered_data = ltv_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика LTV пользователей на {}-й день'.format(horizon))

    # третий график — динамика cac
    ax3 = plt.subplot(2, 3, 3, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in cac_history.index.names if name not in ['dt']]
    filtered_data = cac_history.pivot_table(
        index='dt', columns=columns, values='cac', aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика стоимости привлечения пользователей')

    # четвёртый график — кривые roi
    ax4 = plt.subplot(2, 3, 4)
    roi.T.plot(grid=True, ax=ax4)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('ROI')

    # пятый график — динамика roi
    ax5 = plt.subplot(2, 3, 5, sharey=ax4)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in roi_history.index.names if name not in ['dt']]
    filtered_data = roi_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax5)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.xlabel('Дата привлечения')
    plt.title('Динамика ROI пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show()
```

### Функция для изучения пользовательских данных

Напишем функцию, которая будет принимать название столбца с интересующей нас характеристикой, а возвращать сводную таблицу с данными о:
- значениях признака,
- количестве платящих и неплатящих пользователей,
- долей всех пользователей,
- долей платящих пользоватетей.

<a id='look_into'></a>

```python
def look_into(feature, df):
    print('Анализ пользователей в разрезе', feature)

    # построим сводную таблицу
    pivot = df.pivot_table(index=feature, columns='payer', values = 'user_id', aggfunc='count')\
                    .sort_values(by=True, ascending=False)\
                    .reset_index()
                    
    # сделаем названия столбцов более информативными
    pivot.columns=[feature, 'payer=False', 'payer=True']

    # добавим столбец с общим количеством пользователей
    pivot['total'] = pivot['payer=False'] + pivot['payer=True']
        
    # добавим столбцы с долями в % - долю всех клиентов страны в общем количестве клиентов 
    # и долю платящих клиентов для каждой страны:
    pivot['share_total,%'] = (round(
                        pivot['total'] / 
                        pivot['total'].sum()
                         * 100))
    pivot['share_payer,%'] = (round(
                        pivot['payer=True'] / 
                        pivot['total']
                         * 100))
    return pivot
```

```python
print('Общая сумма расходов на маркетинг:', round(profiles.acquisition_cost.sum()))

# исключаем бесплатных пользователей
profiles_paid = profiles.query('channel != "organic"') 

print('Средние расходы на привлечение 1 пользоватлеля по проекту в целом:', 
      round(profiles.acquisition_cost.sum() / profiles_paid.user_id.nunique(), 2))

# рассчитаем сумму расходов для каждого источника
costs_channel = profiles.pivot_table(index='channel', values=['acquisition_cost', 'user_id'], 
                                     aggfunc={'acquisition_cost':'sum','user_id':'nunique'})\
                        .reset_index()
#  переименуем столбцы
costs_channel = costs_channel.rename(columns={'acquisition_cost':'total_costs', 'user_id':'total_users'})

# добавим столбцы с долей каждого источника в общих расходах, а также с суммой расходов на 1 пользователя
costs_channel['total_costs_share,%'] = round(costs_channel['total_costs'] / costs_channel['total_costs'].sum() * 100)
costs_channel['costs_per_user'] = round(costs_channel['total_costs'] / costs_channel['total_users'], 2)

# отсортируем таблиицу по расходам на 1 пользователя
costs_channel = costs_channel.sort_values(by='costs_per_user', ascending=False)
costs_channel
```

Исследуем подробнее причины различий между США и Европой.   
Проверим гипотезу: пользователи США убыточны, потому что в США преобладают "проблемные" устройства и "проблемные" каналы привлечения. 

```python
# отфильтруем платящих неорганических пользоватлей из США
profiles_usa = profiles_paid.query('payer == True and region == "United States"')

# напишем функцию для расчета количества платящих пользователей и их доли
def look_into2(feature, df, reg):
    print(f'Доля платящих пользователей для {reg} в разрезе {feature}')
    pivot = df.pivot_table(index=feature, values='user_id', aggfunc='nunique')\
                        .sort_values(by='user_id', ascending=False)
    pivot.columns=['users_payers']
    pivot['share,%'] = round(pivot['users_payers']/pivot['users_payers'].sum() * 100)
    return pivot

# теперь вызовем новую функцию, чтобы посмотреть на пользователей из США в разрезе устройств и рекламных каналов
display(look_into2('device', profiles_usa, 'США'))
look_into2('channel', profiles_usa, 'США')
```

В США 68% платящих пользователей предпочитают iPhone и Mac (самые "проблемные" устройства) и 89% пользователей приходят из самых "проблемных" каналов  - FaceBoom и TipTop.

А теперь для сравнения посмотрим на Европу:

```python
# отфильтруем платящих неорганических пользоватлей из Европы
profiles_eur = profiles_paid.query('payer == True and region != "United States"')

# посмотрим на пользователей из Европы в разрезе устройств и рекламных каналов
display(look_into2('device', profiles_eur, 'Европы'))
look_into2('channel', profiles_eur, 'Европы')
```

В Европе полностью отсутствуют самые "проблемные" каналы продвижения - FaceBoom и TipTop. Среди лидеров у них только один убыточный канал - AdNonSense, его доля 27%, что само по себе немало, но существенно ниже доли убыточных каналов (89%) в США. 

Доля самых "проблемных" устройств - iPhone и Mac - в Европе минимальна (30%). Европейцы предпочитают РС.
